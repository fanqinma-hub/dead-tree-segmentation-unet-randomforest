```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

# 
img = cv2.imread("image1.jpg", cv2.IMREAD_GRAYSCALE)

#  Laplacian 
laplacian = cv2.Laplacian(img, cv2.CV_64F)
laplacian = cv2.convertScaleAbs(laplacian)  #  0-255 uint8 

# 
plt.subplot(1, 2, 1)
plt.title("Original")
plt.imshow(img, cmap='gray')

plt.subplot(1, 2, 2)
plt.title("Laplacian")
plt.imshow(laplacian, cmap='gray')

plt.show()
```
